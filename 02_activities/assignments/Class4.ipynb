```python
#Import all libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO


```

```python
# first make our sample data

x= np.array(['Luffy','Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110,180,240,99,220])
y2 = np.array ([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x, y1, color = 'indigo')
ax.plot (x, y2, color = 'red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize= (7,3))
ax.plot(x, y2, color ='red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize= (7,3))
ax.plot(x, y2, color ='red')

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = 'none')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize= (7,3))
ax.plot(x, y2, color ='red')

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor = 'indigo',
            elinewidth = 4,
            capsize = 6,
            capthick = 4)
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')
ax_image = fig.add_axes([0.1,0.11,0.15,0.35])
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')
ax_image = fig.add_axes([0.1,0.11,0.15,0.35])

ax_image.imshow(image)
ax_image.axis('off')
```
