```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x= np.arange (50)
y1 = np.random.randint (0, 100, 50)
y2 = np.random.randint (0, 100, 50)
```

```python
tips = sns.load_dataset ("tips")
print (tips)
```

```python
sns.lineplot(data=tips,
             x = 'total_bill',
             y= 'tip')
sns.set_style('whitegrid')
```

```python
sns.set_style('whitegrid')
```

```python
tipgrpah = sns.lineplot(data=tips,
             x = 'total_bill',
             y= 'tip')

tipgrpah.set (title = 'Tips vs. Total Bill',
              xlabel= 'Total Bill ($)',
              ylabel='Tip Amount ($)')
```

```python

fig = plt.subplots (figsize=(10,3))
tipgrpah = sns.lineplot(data=tips,
             x = 'total_bill',
             y= 'tip')

```

```python
fig = plt.subplots (figsize=(10,3))
tipgrpah = sns.lineplot(data=tips,
             x = 'total_bill',
             y= 'tip',
              color = 'hotpink',
              linestyle = '--',
              linewidth = 3,
              marker = 'o',
              markerfacecolor = 'indigo')

```

```python
fig = plt.subplots (figsize=(10,3))
tipgrpah = sns.scatterplot (data=tips,
             x = 'total_bill',
             y= 'tip',
             style = 'time', 
             hue = 'day',
             palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgrpah.set (title = 'Tips vs. Total Bill',
              xlabel= 'Total Bill ($)',
              ylabel='Tip Amount ($)')
```

```python
sns.pairplot ( data = tips, hue = 'day')
```

```python
daysplot = sns.relplot (data=tips,
           x="total_bill",
           y="tip",
           hue="sex",
           col="day",
           kind="scatter",
           col_wrap=2)
```
