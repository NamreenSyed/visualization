```python
#Import all libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
# Choose random data parameters 
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# Generate scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# Using the same dataset, generate a different plot type e.g. Bar Graph
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

```

```python
# Afix data parameters
font1 = {'family': 'sans-serif','color': 'blue,', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("Total growth over time", fontdict = font1)
ax.set_ylabel("Total growth", fontdict = font2)
ax.set_xlabel("Years since start", fontdict = font2)
```

```python
# Experiment with and add more features to the line graph to improve data visualizations and to dusplay your knowledge of data visualization techniques

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid( axis = "y")
```

```python
# Using the techniques displayed in the previous code, produce a different type of visual to display as percentage of a whole

fig, ax = plt.subplots(figsize=(5,3))

labels = ['A', 'B', 'C', 'D', 'E']
sizes = [25, 15, 35, 10, 15]
colors = ['#7425b9', '#e45d81', '#f1c27d', '#4f98bc', '#F8A74C']
explode = (0.1, 0, 0, 0, 0)

ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', 
       shadow=True, startangle=90, wedgeprops={'edgecolor': '#FA9359', 'linewidth': 2, 'linestyle': '--'})

ax.set_title('Pie Chart?', fontsize=14)
```

```python
from pprint import pprint as pp
import circlify as circ
```

```python
data = [
        0.05, {'id': 'a2', 'datum': 0.05},
        {'id': 'a0', 'datum': 0.8, 'children': [0.3, 0.2, 0.2, 0.1], },
        {'id': 'a1', 'datum': 0.1, 'children': [
            {'id': 'a1_1', 'datum': 0.05}, {'datum': 0.04}, 0.01],
        },
    ]
```

```python
circle = circ.circlify(data, show_enclosure=True)
```

```python
pp(circle)
```

```python

fig, ax = plt.subplots(figsize=(8, 8))

for circle in circles:
    ax.add_patch(plt.Circle((circle['x'], circle['y']), circle['r'], 
                            edgecolor='black', facecolor='skyblue', alpha=0.5))
    
    ax.text(circle['x'], circle['y'], circle['label'], 
            horizontalalignment='center', verticalalignment='center', fontsize=12)
    
plt.show()
```
