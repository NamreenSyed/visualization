```python
#Import all libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

# Generate random values

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# Generate scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1)
ax.plot(x, y2)
```

```python
# Label both lines
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = 'Person1')
ax.plot(x, y2, label = 'Person2')
ax.legend(loc= "lower right")
```

```python
# Modify appearance of legend by adding arguments

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = 'Person1')
ax.plot(x, y2, label = 'Person2')
ax.legend(loc= "lower right",
            frameon = True,
            fontsize = 12,
            ncol = 2,
            shadow = True)
```

```python
# Make legend so it doesn't overlap the graph and data

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = 'Person1')
ax.plot(x, y2, label = 'Person2')
ax.legend(loc= "upper left",
          bbox_to_anchor = (1, 1))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person1')
ax.scatter(x, y2, label = 'Person2')
ax.legend(loc= "lower right")
ax.text(10,95, "This Value is Important!")
```

```python
# Make modifications to identify the 10,95 spot)

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person1')
ax.scatter(x, y2, label = 'Person2')
ax.legend(loc= "lower right")
ax.text(10,95, "This Value is Important!",
        ha = 'center',
        color = 'red',
        size = 20)
```

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform=ax.transData)
ax.text(0.2,0.2, ". Figure: (0.2,0.2)", transform=fig.transFigure)
ax.text(0.5,0.1, ". Axes: (0.5,0.1)", transform=ax.transAxes)
```

```python
# Add an arrow

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc= "lower right")

ax.annotate("This is important!", xy= (10,95),
            xytext=(20,94),
            arrowprops=dict(facecolor = 'black', shrink=0.05))
```

```python
# Add an arrow

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc= "lower right")

ax.annotate("This is important!", xy= (10,95),
            xytext=(20,94),
            arrowprops=dict(facecolor = 'hotpink', arrowstyle = "wedge"))
```

```python
# Remove ticks and labels

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc= "lower right")

ax.yaxis.set_major_locator(plt.NullLocator()) # remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) #Remove only labels
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc= "lower right")

ax.yaxis.set_major_locator(plt.MaxNLocator(3)) 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc= "lower right")

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) 
```

```python
#Import all libraries for NEW TOPIC

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# Generate random values

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.random.randint([110,180,240,99,220])
```

```python


fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1, 
                               figsize =(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# Activity: Customize our plots


fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1, 
                               figsize =(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)

ax1.scatter(x1, y1, label = 'TTC', marker = '*')
ax1.legend(loc= "upper right")


ax2.bar(x2,y2,
           color = '#7425b9')

```

```python
fig, someaxes = plt.subplot_mosaic ([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize=(7,4))

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic ([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize=(7,4),
                                     layout= 'constrained')

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel("A big label", fontsize =18)
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```
